```python
import pandas as pd

# Sample DataFrames
df1 = pd.DataFrame({
    'id': [1, 2, 3],
    'name': ['Alice', 'Bob', 'Charlie']
})

df2 = pd.DataFrame({
    'id': [2, 3, 4],
    'age': [24, 25, 26]
})

# Merging the DataFrames on 'id'
merged_df = pd.merge(df1, df2, on='id', how='inner')
print(merged_df)
```
